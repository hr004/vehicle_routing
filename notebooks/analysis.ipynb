```python
import seaborn as sns
import pandas as pd
```

### Task 1

solver default runtime vs predicted configuration's runtime

```python
runs = pd.read_csv("./../artifacts/n_instances/runs_average.csv")
runs
```

```python
ax = sns.lineplot(runs[["predicted_runtime", "solver_default_runtime"]])
ax.set_xlabel("n_instances")
ax.set_ylabel("runtime")
ax.set_xticks(range(len(runs['n_train_instances'])))
ax.set_xticklabels(runs['n_train_instances'])
```

Performance comparison (Improvement in runtime)

```python
runs['improvement'] = runs['solver_default_runtime'] - runs['predicted_runtime']
sns.barplot(x = runs['n_train_instances'], y = runs['improvement'])
```

```python
parameters = [p for p in runs.columns if p.startswith('param_')]
runs[['n_train_instances', 'predicted_runtime', 'solver_default_runtime'] + parameters]
```

#### Comparing Random Forest, Gaussian Process and Gradient Boost

```python
rf = pd.read_csv("./../artifacts/n_locations/runs_average.csv")
gp = pd.read_csv("./../artifacts/n_locations_gaussian_process_regressor/runs_average.csv")
gb = pd.read_csv("./../artifacts/n_locations_gradient_boosting_regressor/runs_average.csv")
```

```python
def add_improvement_row(df):
    df['improvement'] = df['solver_default_runtime'] - df['predicted_runtime']
    return df

rf = add_improvement_row(rf)
gp = add_improvement_row(gp)
gb = add_improvement_row(gb)
```

```python
gp
```

```python
gb
```

```python
comparison_df = pd.concat([rf['improvement'], gb['improvement'], gp['improvement']], axis = 1, keys = ['rf', 'gb', 'gp'])
```

```python
ax = sns.violinplot(comparison_df)
ax.set_ylabel("improvement")
```
