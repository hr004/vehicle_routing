### Question 1
For this task, I kept testing instances to a constant value of 20 while varying training instances from 3 to 35. For each value of n_instances, value runtime is taken as target variable. For the first run, it can be seen that the gap between solver's default configuration runtime and our predicted configuration runtime is higher. As we increase n_instances, the gap starts to decrease as the model generalizes better with increasing number of instances. An important observation is that, there are some sudden jumps in predicted configuration's runtime. This might be due to taking average runtime as target variable for `n` number of instances. 

Once we change our target variable to more robust statistic such as percentile (I took 75th percentile of n_instances runtimes), the model starts to generalize better than average runtime as target variable. As we took only 50 samples randomly, it is possible that there's bias in the training data and hence we observe some variation in the actual and predicted runtime.

```python
runs = pd.read_csv("./artifacts/n_instances/runs_average.csv")
ax = sns.lineplot(runs[["default_test_runtime", "predicted_test_runtime"]])
ax.set_xlabel("n_instances")
ax.set_ylabel("runtime")
```

```python
runs = pd.read_csv("./artifacts/n_instances/runs_percentile.csv")
ax = sns.lineplot(runs[["default_test_runtime", "predicted_test_runtime"]])
ax.set_xlabel("n_instances")
ax.set_ylabel("runtime")
```
